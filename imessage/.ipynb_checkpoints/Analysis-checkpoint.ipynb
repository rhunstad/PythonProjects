# Individual Analysis

# Functions used

```python
def total_by_group(group):
    j = convo.groupby(['is_from_me', group])['text'].count().reset_index().groupby('is_from_me').get_group(0).drop('is_from_me', axis=1)
    r = convo.groupby(['is_from_me', group])['text'].count().reset_index().groupby('is_from_me').get_group(1).drop('is_from_me', axis=1)
    hours = j.merge(r, how='outer', on=group, suffixes=['_j', '_r']).fillna(0)

    X = hours[group]
    j_hour = hours['text_j']
    r_hour = hours['text_r']

    X_axis = np.arange(len(X))
    X_label = X
    
    plt.figure(figsize=(12, 6), dpi=250)
    plt.bar(X_axis - 0.2, j_hour, 0.4, label = 'Justin', color=justin_color)
    plt.bar(X_axis + 0.2, r_hour, 0.4, label = 'Ryland', color=ryland_color)

    if group == "hour":
        X_label = ['12am', '1am', '2am', '3am', '4am', '5am', '6am', '7am', '8am', '9am', '10am', '11am', '12pm', '1pm', '2pm', '3pm', '4pm', '5pm', '6pm', '7pm', '8pm', '9pm', '10pm', '11pm']
    elif group == 'weekday':
        X_label = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturady', 'Sunday']
    elif group == 'month':
        X_label = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'][0:len(X)]
    
    plt.xticks(X_axis, X_label, rotation=45)
    plt.xlabel(group)
    plt.ylabel("Number of Texts")
    plt.title(f"Texts by {group}")
    plt.legend()
    plt.show()
    
    
def get_words(from_me): 
    messages = list(convo[(convo['reaction'].isna()) & (convo['is_text'] == 1) & (convo['is_from_me'] == from_me)]['text'])
    x = " ".join(messages)
    x = re.sub('[^a-zA-Z\s]', '', x)
    x = re.sub('\n', ' ', x)
    x = re.sub('\s+', ' ', x)
    x = x.strip()
    x = x.lower()
    corpus = x.split(" ")
    words = pd.DataFrame(data={"words": corpus})
    return words


def word_count(from_me):
    if from_me:
        name = "Ryland"
        col = ryland_color
    else:
        name = "Justin"
        col = justin_color

    words = get_words(from_me)
    unique_words = words['words'].unique()
    num_words = words.shape[0]
    num_unique = unique_words.shape[0]
    percent = round(num_unique*100/num_words,2)

    print(f"{name}: \n - Total Words: {num_words:,}\n - Total Unique: {num_unique:,} \n - Ratio: {round(100/percent, 2)} ({percent}%)\n")
    
    text = " ".join(list(words['words']))
    cloud = WordCloud(max_font_size=50, max_words=50, background_color=col).generate(text)
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

    
def zipf(from_me):
    max_index = 150
    words = get_words(from_me)
    count_df = words.groupby('words')['words'].count().rename("count").reset_index().sort_values('count', ascending=False).reset_index().drop("index", axis=1)
    most = count_df.iloc[0]['count']
    count_df['freq'] = count_df['count'].map(lambda x: round(x/most, 6))
    count_df['zipf'] = count_df.index.map(lambda x: round(1/(x+1), 6))
    count_graph = count_df[count_df.index < max_index]
    return count_graph


```
